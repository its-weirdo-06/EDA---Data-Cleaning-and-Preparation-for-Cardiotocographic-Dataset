```python
# Outlier Detection: User defined Function to calculate Upper Extreme and Lower Extreme value
def extreme_detection(data,colname):
  q1 = data[colname].quantile(0.25)
  q3 = data[colname].quantile(0.75)
  iqr = q3 - q1

  upper_extreme = q3 + (1.5 * iqr)
  lower_extreme = q1 - (1.5 * iqr)

  return lower_extreme, upper_extreme
```

```python
for i in df.columns:
    print(i+':',extreme_detection(df,i))
```

```python
for i in df.columns:
    lower_extreme= extreme_detection(df,i)[0]
    upper_extreme= extreme_detection(df,i)[1]
    df.loc[df[i]<lower_extreme,i]= lower_extreme
    df.loc[df[i]>upper_extreme,i]= upper_extreme
```

```python
stats_summary=df.describe()
IQR ={}
for i in df.columns:
    IQR[i]=df[i].quantile(0.75)- df[i].quantile(0.25)
IQR_df =pd.DataFrame(IQR,index=['IQR'])
stats_summary=pd.concat([stats_summary,IQR_df])
stats_summary
```

```python
# Visualization for LB - Baseline Fetal Heart Rate (FHR) using Histogram(seaborn)
plt.figure(figsize=(10, 6))
sns.histplot(df['LB'], bins=20, kde=True)
plt.title('Histogram of Baseline Fetal Heart Rate (FHR)')
plt.xlabel('Baseline Fetal Heart Rate (FHR)')
plt.ylabel('Frequency')
plt.show()
```

```python
vars_to_plot = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV']
```

```python
# Calculating the correlation matrix
correlation_matrix = df[vars_to_plot].corr()

# Creating the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Creating violin plots for each variable
plt.figure(figsize=(12, 8))
for i, var in enumerate(vars_to_plot):
    plt.subplot(4, 3, i + 1)
    sns.violinplot(x=df[var])
    plt.title(var)
plt.tight_layout()
plt.show()
```
